```python
import pandas as pd
from pathlib import Path

# Đường dẫn tập dữ liệu
DATA_PATH = Path("dataset/heart.csv")

# Đọc dữ liệu
df = pd.read_csv(DATA_PATH)
```

## Nguồn dữ liệu

```python
nguon_du_lieu = {
    "Tên tập dữ liệu": "Heart Disease Dataset",
    "Nguồn": "UCI Machine Learning Repository / Kaggle (johnsmith88)",
    "URL": "https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset",
    "Mô tả": (
        "Dữ liệu từ năm 1988, tổng hợp từ 4 cơ sở: Cleveland, Hungary, "
        "Switzerland và Long Beach V. Gồm 76 thuộc tính gốc; các nghiên cứu "
        "công bố chỉ dùng một tập con gồm 14 thuộc tính. Trường 'target' cho "
        "biết tình trạng mắc bệnh tim: 0 = không bệnh, 1 = có bệnh."
    ),
    # Dùng đường dẫn tương đối so với gốc dự án: không lộ cấu trúc thư mục máy
    # cục bộ khi notebook được công bố.
    "File cục bộ": str(DATA_PATH),
}

for key, value in nguon_du_lieu.items():
    print(f"• {key}: {value}")
```

## Số lượng mẫu

```python
so_mau = len(df)

print(f"• Số lượng mẫu (số dòng): {so_mau:,}")
print(f"• Số thuộc tính (số cột): {df.shape[1]}")
print(f"• Kích thước tập dữ liệu: {df.shape[0]:,} dòng × {df.shape[1]} cột")

# Lưu ý: bản heart.csv (1025 dòng) là phiên bản nhân bản từ dữ liệu gốc
# Cleveland (303 mẫu). Điều này không ảnh hưởng tới quy trình kiểm định minh họa.
print(f"\n• Số dòng trùng lặp: {df.duplicated().sum():,}")
```

## Các thuộc tính

```python
print("Danh sách các thuộc tính:")
for i, col in enumerate(df.columns, start=1):
    print(f"  {i}. {col}")
```

## Kiểu dữ liệu của từng thuộc tính
